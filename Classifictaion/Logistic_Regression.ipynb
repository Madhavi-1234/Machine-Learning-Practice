```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= df[df['species']!='setosa']
df
```

```python
df['species']= df['species'].map({'versicolor':0,'virginica':1}) # to map the species tnto 0 and 1
df.head()
```

```python
print(df['species'].unique())
```

```python
# split the dataset into dependant and independant features

x= df.iloc[:,:-1]
y= df.iloc[:,-1]
```

```python
# split train test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=2)
```

```python
# model
from sklearn.linear_model import LogisticRegression

classifier= LogisticRegression()
```

```python
# hyperperameter tuning GridSearchCV

from sklearn.model_selection import GridSearchCV
parameter= {'penalty':['l1','l2','elasticnet'],'max_iter':[100,200,300],'C':[1,2,3,4,5,10,20,30,40]}
```

```python
classifier_reg= GridSearchCV(classifier, parameter, scoring= 'accuracy',cv= 5)
```

```python

classifier_reg.fit(x_train, y_train)

print('Best accuracy score: ', classifier_reg.best_score_)
```

```python
classifier_reg.best_params_
```

```python
# predict

y_pred= classifier_reg.predict(x_test)
```

```python
## accuracy

from sklearn.metrics import accuracy_score, classification_report

score= accuracy_score(y_test, y_pred)
score
```

```python
print(classification_report(y_test, y_pred))
```
