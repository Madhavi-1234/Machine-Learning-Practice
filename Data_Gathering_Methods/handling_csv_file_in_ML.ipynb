## use_ header parameter

```python
pd.read_csv('file.csv', header=1)
```

## Skip rows/nrows paramter

```python
#skip the mentioned rows
pd.read_csv('file.csv', skiprows=[0,1])

# nrows
pd.read_csv('file.csv', nrows=100)
```

## Encoding paramter

```python
## encoding param helps when loading file with different encoding
pd.read_csv('file.csv', encoding='latin-1'). # will transform into normla dataframe by specifying the encoding value
```

## Skip the bad lines

```python
# where the data is inappropiate we use this function

pd.read_csv('file.csv', error_bad_lines=False) # false will skip the line where data was not appropriate
```
