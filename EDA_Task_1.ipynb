```python
import numpy as np
import matplotlib.pyplot as plt

# X (features/pixels) or y (labels)

X = df.drop('label', axis=1).values   #  pixel columns
y = df['label'].values                  # only label column

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

# 28x28 grid reshape
first_image = X[0].reshape(28, 28)
plt.figure(figsize=(4, 4))
plt.title(class_names[y[0]])
plt.imshow(first_image, cmap='gray', interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
class_counts = df['label'].value_counts().sort_index()

print("Class Distribution:")
for label, count in class_counts.items():
    print(f"{label}- {class_names[label]:<12}: {count} images")
```

```python
plt.figure(figsize=(10, 5))
plt.bar(class_counts.index, class_counts.values, color='skyblue', edgecolor='black')
plt.xticks(class_counts.index, class_names, rotation=45, ha='left')
plt.title("Class Distribution - Fashion MNIST")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()
```

```python
# Saare pixel values ek saath except the label column
all_pixels = X.flatten()

plt.figure(figsize=(8, 5))
plt.hist(all_pixels, bins= 50, color= "skyblue", edgecolor= "black")
plt.title("Pixel Intensity Distribution (All Images)")
plt.xlabel("Pixel Value (0-255)")
plt.ylabel("Frequency")
plt.show()

print("Mean pixel value:", all_pixels.mean())
print("Median pixel value:", np.median(all_pixels))
```

```python
fig, axes = plt.subplots(1, 10, figsize=(20, 3))  # 1 row, 10 columns
fig.suptitle("Average Image per Class", fontsize=14)

for i in range(10):
    class_images = X[y == i]
    avg_image = class_images.mean(axis=0).reshape(28, 28)

    axes[i].imshow(avg_image, cmap='gray', interpolation='nearest')
    axes[i].set_title(class_names[i], fontsize=8)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
# Har class ka average pixel vector nikalo
class_avg_vectors=[]
for i in range(10):
    avg = X[y==i].mean(axis=0)
    class_avg_vectors.append(avg)

class_avg_vectors = np.array(class_avg_vectors)
print(class_avg_vectors.shape)
```

```python
similarity_matrix = np.corrcoef(class_avg_vectors)
import pandas as pd

# Similarity matrix ko DataFrame mein convert karo with labels
sim_df = pd.DataFrame(similarity_matrix, index=class_names, columns=class_names)

# Round karo 2 decimal places tak (clean dikhega)
sim_df = sim_df.round(2)

print(sim_df)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(similarity_matrix, cmap='RdYlGn', vmin=-1, vmax=1)


plt.xticks(range(10), class_names, rotation=45, ha='right')
plt.yticks(range(10), class_names)

plt.colorbar(label='Similarity Score')

for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{similarity_matrix[i,j]:.2f}',ha='center', va='center', fontsize=8)

plt.title("Class Similarity Heatmap")
plt.tight_layout()
plt.show()
```

```python
# pca
# dimensionality reduction and for better visulaization
from sklearn.decomposition import PCA

# 784 dimensions ko 2 dimensions mein reduce
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

print("Original shape:", X.shape)
print("After PCA shape:", X_pca.shape)
print("Variance explained:", pca.explained_variance_ratio_)
```
